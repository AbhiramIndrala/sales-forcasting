```python
wall_mart.to_csv("wallmart.csv",index=False)
df=pd.read_csv("wallmart.csv")
df.head()
```

```python
print(df.shape)#Shape of the dataset
print(df.size)#Size of the dataset
print(df.info())
```

```python
df.isnull().sum()
df.shape
```

```python
#Modiying the dataset
df.replace({"IsHoliday":{False:0,True:1}},inplace=True)
df.sample(5)
```

```python
#Handling categorical data
df["Type"].value_counts()
```

```python
type_df=pd.get_dummies(df["Type"],drop_first=True,prefix="Type")
```

```python
type_df.head(2)
```

```python
df=pd.concat([df,type_df],axis=1)
```

```python
df.head(3)
```

```python
df.drop("Type",axis=1,inplace=True)
```

```python
df.head(2)
```

```python
#Handling dates
#Handling the date and time
df.Date.dtype
date=pd.to_datetime(df["Date"])
date.dt.month
date.dt.year
date.dt.day
df["Month"]=date.dt.month
```

```python
df.drop("Date",axis=1,inplace=True)
```

```python
df.head(2)
```

```python
#Handling the null values
df.isnull().sum()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(df.isna())
plt.show()
```

```python
df.fillna(value=0,inplace=True)
df.isnull().sum()
```

```python
df.drop(["CPI","Unemployment","Fuel_Price"],axis=1,inplace=True)
```

```python
len(df.columns)
```

```python
x=df.drop(["Weekly_Sales"],axis=1).values
```

```python
x.shape
```

```python
y=df["Weekly_Sales"].values
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mse
```

```python
np.sqrt(mse)
```

```python
#SGD
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cv_score=cross_val_score(model,x_train,y_train,n_jobs=4,cv=5,scoring="neg_mean_squared_error")
```

```python
np.sqrt(-(cv_score))
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
sgd=SGDRegressor()
```

```python
sgd.fit(x_train,y_train)
```

```python
y_pred2=sgd.predict(x_test)
```

```python
sgd_error=mean_squared_error(y_test,y_pred2)
```

```python
sgd_error
```

```python
np.sqrt(sgd_error)
```
